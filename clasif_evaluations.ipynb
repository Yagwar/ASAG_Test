# General and datasets Scores
## Train

```python
train_predictions=train_answers[["KNN","RBF_SVM","Rand_Forest","AdaBoost","Naive_Bayes","QDA"]]
train_predictions
```

```python
gral_scores_train=[f1_score(train_answers.st_ans_calif, train_predictions[prediction], average='weighted') for prediction in train_predictions]
gral_scores_train.extend([np.nan,np.nan])
```

```python
SciEntsBank_scores_train=[f1_score(train_answers.st_ans_calif[train_answers.dtset=="TRAIN SciEntsBank"], train_predictions.loc[train_answers.dtset=="TRAIN SciEntsBank",prediction], average='weighted') for prediction in train_predictions]
SciEntsBank_scores_train.extend([np.nan,np.nan])
```

```python
Beetle_scores_train=[f1_score(train_answers.st_ans_calif[train_answers.dtset=="TRAIN Beetle"], train_predictions.loc[train_answers.dtset=="TRAIN Beetle",prediction], average='weighted') for prediction in train_predictions]
Beetle_scores_train.extend([np.nan,np.nan])
```
